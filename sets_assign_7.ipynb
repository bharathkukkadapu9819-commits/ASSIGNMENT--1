```python
print("Question 1: Create a set of first 10 natural numbers")
natural_numbers=set([i for i in range(1,11)])
natural_numbers
```

```python
print("Question 2: Add element 11 to set {1, 2, 3, 4, 5}")
set_elements={1, 2, 3, 4, 5}
set_elements.add(11)
set_elements
```

```python
print("Question 3: Remove element 3 from set {1, 2, 3, 4, 5, 6}")
removing_elements={1, 2, 3, 4, 5, 6}
removing_elements.discard(3)
removing_elements
```

```python
print("Question 4: Find the intersection of {1, 2, 3, 4, 5} and {4, 5, 6, 7, 8}")
set1={1, 2, 3, 4, 5}
set2={4, 5, 6, 7, 8}
set1.intersection(set2)
# set1 & (set2)
```

```python
print("Question 5: Find the difference between {1, 2, 3, 4, 5} and {4, 5, 6, 7, 8}")
set3={1, 2, 3, 4, 5}
set4={4, 5, 6, 7, 8}
set3.difference(set4)
# set3-set4
```
